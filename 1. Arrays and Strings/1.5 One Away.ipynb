**1.5 One Away:** There are three types of edits that can be performed on strings: insert a character,
remove a character, or replace a character. Given two strings, write a function to check if they are
one edit (or zero edits) away. 

This is one of those problems where it's helpful to think about the "meaning" of each of these operations.

*What does it mean for two strings to be one insertion, replacement, or removal away from each other?*


**• Replacement:** Consider two strings, such as 'bale' and 'pale', that are one replacement away. Yes, that does mean that you could replace a character in 'bale' to make 'pale'. But more precisely, it means that
they are different only in one place.

**• Insertion:** The strings 'apple' and 'aple' are one insertion away. This means that if you compared the strings, they would be identical—except for a shift at some point in the strings.

**• Removal:** The strings 'apple' and 'aple' are also one removal away, since removal is just the inverse of insertion. 

```

def solution_1(string1,string2):

  if len(string1)==len(string2):
    return replacement(string1,string2)
  
  elif abs(len(string1)-len(string2))==1:
    return insertion_removal(string1,string2)
  else:
    return False
```

```
def replacement(string1,string2):
  s = set()
  for ch in string1:
    s.add(ch)
  for ch in string2:
    if ch in s:
      s.remove(ch)
    else:
      s.add(ch)
  return len(s)==1 or len(s)==0
```

```
def insertion_removal(string1,string2):
  s = set()

  if len(string1)>len(string2):
    longer=string1
    smaller=string2
  else:
    longer=string2
    smaller=string1
    
  for ch in longer:
    s.add(ch)
  for ch in smaller:
    if ch in s:
      s.remove(ch)
    else:
      s.add(ch)
  
  return len(s)==1
```

```
solution_1('abc','abc')
```

```
solution_1('abc','abcd')
```
